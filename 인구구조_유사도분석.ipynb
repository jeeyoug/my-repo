```python
import csv
import numpy as np

f = open('age.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
data1 = list(data)


name = input('인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요:')
min = 1
result_name = ''
result = 0

for row in data1:
  if name in row[0] :
    home = np.array(row[3:], dtype = int) / int(row[2])
  for row in data1 : 
    away = np.array(row[3 :], dtype = int) / int(row[2])
    s = (np.sum(home - away))
    if s < min :
      min = s
      result_name = row[0]
      result = away

  
plt.rc('font', family= 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title(name + '지역의 인구 구조')
plt.plot(home)
plt.plot(result)
plt.show()
```

```python
import csv
import numpy as np

f = open('age.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
data1 = list(data)


name = input('인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요:')
min = 1
result_name = ''
result = 0

for row in data1:
  if name in row[0] :
    home = np.array(row[3:], dtype = int) / int(row[2])
  for row in data1 : 
    away = np.array(row[3 :], dtype = int) / int(row[2])
    s = (np.sum(home - away) ** 2)
    if s < min and name not in row[0] :
      min = s
      result_name = row[0]
      result = away

  
plt.rc('font', family= 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title(name + '지역의 인구 구조')
plt.plot(home, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
